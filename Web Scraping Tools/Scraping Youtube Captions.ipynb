```python
from youtube_transcript_api import YouTubeTranscriptApi
import re
# assigning srt variable with the list
# of dictionaries obtained by the get_transcript() function
srt = YouTubeTranscriptApi.get_transcript("ULTu5x5C8h8")
# prints the result
```

```python
DOCUMENTARYLIST=[]
LINKSWORKED=[]
for i,j in zip(LINK_CLEANED,LINK):
  try:
    srt= YouTubeTranscriptApi.get_transcript(i)
    JOIN=""
    TEMPLIST=[]
    for i in srt:
      i=str(i)
      RE=re.findall(r"(?<=\'text\'\: \').*(?=\'\, \'start\'.+\b)",i)
      RE=str(RE)
      #print(RE,"RE")
      RE2=re.sub(r"\\n",r" ",RE)
      RE3=re.sub(r"\\",r"",RE2)
      RE4=re.sub(r"\[\'",r"",RE3)
      RE5=re.sub(r"\']",r"",RE4)
      RE6=re.sub(r"\[\]",r"",RE5)
      print(RE6)
      ## we don't remove musical notes or [music] as we can use this information for text classification. This would be done during preprocess for the actual synthesis
      #print(RE6)
      TEMPLIST.append(RE6)
    #print(TEMPLIST)
    for i in TEMPLIST:
      print(len(i))
      #print(i, "I")
      if len(i)>0:
        JOIN=JOIN+" "+i
    JOIN=JOIN[1:]
    print(JOIN[0])
    DOCUMENTARYLIST.append(JOIN)
    LINKSWORKED.append(j)
  except:
    print("Dumbass")
#print(DOCUMENTARYLIST[0])
#print(len(DOCUMENTARYLIST))
```

```python
import pandas as pd
filename=""
Docu_Tag=["Documentary"]*len(DOCUMENTARYLIST)
print(len(DOCUMENTARYLIST))
df = pd.DataFrame(list(zip(Docu_Tag, DOCUMENTARYLIST,LINKSWORKED)),columns =['Genre', 'Text','Link'])
df.to_csv(filename)
```
