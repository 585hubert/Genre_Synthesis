```python
for link in Retrieved_URLs:
  try:
      #time.sleep(0.25)
      status, response = http.request(link)
      soup=BeautifulSoup(response, "html.parser")
      soup2=soup.find_all(class_="cnnBodyText")
      soup3=str(soup2)
      #print(soup3)
      try:
        RE=re.findall(r"(?<=\[\d\d\:\d\d\:\d\d\]\<br\/\>).*(?=\(COMMERCIAL BREAK\).+\b)",soup3)
        REMA=md_to_text(RE[0])
        NEWS_LIST.append(REMA)
        Successful_URLs.append(link)
      except:
        try:
          RE=re.findall(r"(?<=\[\d\d\:\d\d\:\d\d\] ).*(?=\(COMMERCIAL BREAK\).+\b)",soup3)
          REMA=md_to_text(RE[0])
          NEWS_LIST.append(REMA)
          Successful_URLs.append(link)
        except:
          if link == 'https://transcripts.cnn.com/' or link == "https://transcripts.cnn.com/providers\" target=\"providers" or link == "https://transcripts.cnn.com/":
            print("Expected non-existent link")
            pass
          elif re.find("Did Not Air",link) or re.find("did not air", link):
            print("Expected Did Not Air")
            pass
          else:
            print(link,"failed")
            pass
  except:
    pass

```
