```python
def LenListHistogram(ListofLists,Labels,Question,SampleOverride=False,P_Values=False,Name=True):
  #Import some general stylistic features#
  import matplotlib.font_manager
  plt.rcParams['font.family'] = 'cmr10'
  plt.rcParams['mathtext.it'] = 'STIXGeneral:italic'
  matplotlib.rc('text', usetex = False)
  ##
  #Set up the figure#
  fig,ax=plt.subplots(figsize=(12,7),facecolor='#0B0227')
  plt.grid(visible=True,zorder=-1,linewidth=1,color="white",alpha=0.2)
  plt.gca().set_axisbelow(True)
  fig.tight_layout()
  for axis in ['top', 'bottom', 'left', 'right']:
    ax.spines[axis].set_color('#FFFFFF')
  for ticks in ['x','y']:
    ax.tick_params(axis=ticks,color='#FFFFFF',labelcolor='#FFFFFF',width=2)
  ax.set_ylim([0.5, 5])
  ax.xaxis.label.set_color('#FFFFFF')
  ax.yaxis.label.set_color('#FFFFFF')
  ax.spines['left'].set_linewidth(2)
  ax.spines['bottom'].set_linewidth(2)
  ax.spines['right'].set_visible(False)
  ax.spines['top'].set_visible(False)
  ax.set_facecolor("#0B0227")
  plt.xticks(fontsize=15)
  plt.yticks(fontsize=15)
  #Label your axes and do your title#
  ax.set_ylabel('Evaluator Rating',fontdict={'fontsize': '20', 'fontweight': '3'})
  ax.set_xlabel('Samples Compared',fontdict={'fontsize': '20', 'fontweight': '3'})
  TITLE=Question
  ax.set_title(TITLE,color='#FFFFFF', fontdict={'fontsize': '25', 'fontweight': '3'})
  #parse your lists#
  avglist=[]
  stdlist=[]
  Len=[]
  for i in ListofLists:
    try:
      i.dropna(inplace=True)
    except:
      pass
    Len.append(len(i))
    avglist.append(round(np.nanmean(i),3))
    stdlist.append(round(np.nanstd(i),3))
  print(avglist,stdlist)
  #actually plot the thing#
  bars=ax.bar(range(0,len(avglist)),avglist,tick_label=Labels,yerr=stdlist,color="#FF785A",error_kw=dict(ecolor='#A9B4C2', lw=2, capsize=5, capthick=2))
  #
  #bar specific conditionals#
  try:
    a=Labels.index("FastSpeech2")
    bars[a].set_color('#FFF500')
  except:
    pass
  #
  #generating text based on bar location#
  for k,i,j in zip(range(0,len(avglist)),avglist,stdlist):
    percent=round(i,3)
    text=percent
    Textupper=round(percent+j,3)
    Textlower=round(percent-j,3)
    #print(Textupper,Textlower)
    plt.text(k-0.18,i+0.1,text,bbox=dict(facecolor='#f5f5f5'))
    plt.text(k-0.18,i+0.1+j,Textupper,bbox=dict(facecolor='#f5f5f5'))
    plt.text(k-0.18,i+0.1-j,Textlower,bbox=dict(facecolor='#f5f5f5'))
  #
  #this is for when you can't just fetch len()
  if SampleOverride == False:
    TextStr="Responses = "+str(Len[0])
  else:
    number=input("How many samples?")
    TextStr="Responses = "+str(number)
  plt.text((len(avglist)-1),4.5, TextStr, fontsize=12, bbox=dict(facecolor='#86979F'))
  # unnecessarily verbose for this implementation, but worth having#
  for p in ax.patches:
      a=p.get_width()
  #pvalue generator based on externally provided list and bar location#
  if P_Values is not False:
    for i,j in zip(P_Values,range(1,len(P_Values)+1)):
      p_val=round(i,3)
      text=r"$\mathit{"+"p = "+str(p_val)+"}$"
      horiz=0.5*((j-(a/2))+(j+(a/2)))
      vert=0.5*((1-(a/2))+(1+(a/2)))
      print(horiz,vert)
      plt.text(horiz,vert,text,fontsize=15,ha="center",weight="bold",style='italic')
  if Name is True:
    name=input("Name The File:")
    plt.savefig(name,bbox_inches="tight")
  #
  return plt.show()


# use this to test the function
#LenListHistogram(KNNFS2,Samples_KNNFS2,"Q2")
```
