```python
def Accuracy(Matrix):
  import numpy as np
  index=0
  Correct=0
  Total=0
  for i in Matrix:
    Correct=Correct+i[index]
    Total=Total+np.sum(i)
    index+=1
  return Correct/Total
```

```python
def Precision_Recall_F1(Matrix):
  import numpy as np
  indexA=0
  indexB=0
  MatrixT=np.array(Matrix).transpose()
  PrecisionList=[]
  RecallList=[]
  F1List=[]
  for i,j in enumerate(Matrix):
    Recall=j[indexA]/np.sum(j)
    RecallList.append(Recall)
    indexA+=1
  for i,j in enumerate(MatrixT):
    Precision=j[indexB]/np.sum(j)
    PrecisionList.append(Precision)
    indexB+=1
  for i,j in zip(PrecisionList,RecallList):
    F1=2*(i*j)/(i+j)
    F1List.append(F1)
  return PrecisionList, RecallList, F1List

#|* TEST SET *| |ACC| 96.3304 |PRECISION| 0.9441 |RECALL| 0.9073 |F1| 0.9244 #600-100 Test
#[7227  487    4   25][  382 17220     3     8][14  5 77  6][  48   15    3 1727]
Precision,Recall,F1=Precision_Recall_F1([[7227,487,4,25],[382,17220,3,8],[14,5,77,6],[48,15,3,1727]])
#print(np.mean(Precision),np.mean(Recall),np.mean(F1))

```

```python
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
def plot_confusion_matrix(Matrix,Title,ListofTicks,RotateText=True,Normalise=True,LabelOverride=False,SaveFig=False):
  if Normalise==True:
    Div1=sum(Matrix[0]) #for normalisation
    Div2=sum(Matrix[1])
    Div3=sum(Matrix[2])
    Div4=sum(Matrix[3])
    print(Div1,Div2,Div3,Div4)
    DivList=[Div1,Div2,Div3,Div4]
    for a,i in enumerate(Matrix): #to actually normalise the matrix
      for b,j in enumerate(i):
        Matrix[a][b]=round(j/DivList[a],3)
  fig,ax=plt.subplots(figsize=(11,6),facecolor='#0B0227')
  im=plt.imshow(Matrix)
  if LabelOverride is not False:
    ax.set_ylabel(LabelOverride[1],fontdict={'fontsize': '20', 'fontweight': '3'})
    ax.set_xlabel(LabelOverride[0],fontdict={'fontsize': '20', 'fontweight': '3'})
  else:
    ax.set_ylabel('Predicted',fontdict={'fontsize': '20', 'fontweight': '3'})
    ax.set_xlabel('Actual',fontdict={'fontsize': '20', 'fontweight': '3'})
  ax.set_xticks([0,1,2,3])
  ax.set_yticks([0,1,2,3])
  ax.xaxis.label.set_color('#FFFFFF')
  ax.yaxis.label.set_color('#FFFFFF')
  ax.set_title(Title,color='#FFFFFF', fontdict={'fontsize': '25', 'fontweight': '3'})
  for ticks in ['x','y']:
    ax.tick_params(axis=ticks,color='#FFFFFF',labelcolor='#FFFFFF',width=2)
  ax.xaxis.set_minor_locator(MultipleLocator(0.5)) #This is needed for the grid, as it relies on mapping onto minor ticks
  ax.yaxis.set_minor_locator(MultipleLocator(0.5)) #This is needed for the grid, as it relies on mapping onto minor ticks
  cb=plt.colorbar()
  for i in range(len(Matrix[0])):
    for j in range(len(Matrix[0])):
      text = ax.text(j, i, round(Matrix[i][j],3),
                       ha="center", va="center", color="black",fontsize=12,bbox=dict(facecolor='white',alpha=0.4,edgecolor='none'))
  cb.ax.yaxis.set_tick_params(color='#FFFFFF')
  plt.setp(plt.getp(cb.ax.axes, 'yticklabels'), color='#FFFFFF')
  cb.outline.set_edgecolor('#FFFFFF')
  cb.ax.tick_params(labelsize=12)
  ax.grid(which="minor",color='w', linestyle='-', linewidth=1)
  for axis in ['top', 'bottom', 'left', 'right']:
    ax.spines[axis].set_color('#FFFFFF')
  plt.xticks(range(len(ListofTicks)),ListofTicks,fontsize=15)
  if RotateText==True:
    plt.xticks(rotation=30, ha="right")
    plt.yticks(rotation=30, ha="right")
  plt.yticks(range(len(ListofTicks)),ListofTicks,fontsize=15)
  if SaveFig is not False:
    plt.savefig(SaveFig,bbox_inches="tight")
  plt.show()
```
